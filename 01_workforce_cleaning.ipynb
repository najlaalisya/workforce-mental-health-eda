```python
import pandas as pd
import numpy as np
import uuid
import random

def generate_public_synthetic_dataset(rows=15000, seed=42):
    random.seed(seed)
    np.random.seed(seed)

    data = []

    # -----------------------------
    # Define categories
    # -----------------------------
    countries = ["Malaysia", "Indonesia", "Vietnam", "Thailand", "Philippines", "Singapore", "China", "United States"]
    genders = ["Male", "Female", "Non-binary", np.nan]
    employment_statuses = ["Full-time", "Freelance", "Part-time", "Student", "Unemployed", "Retired"]
    work_modes = ["On-site", "Hybrid", "Remote"]
    risk_tiers = ["Low", "Moderate", "High"]

    for _ in range(rows):
        # About 35% complete the core wellbeing modules
        is_completer = random.random() < 0.35

        row = {
            # -----------------------------
            # Identifiers & metadata
            # -----------------------------
            "respondent_id": str(uuid.uuid4())[:8],
            "submission_year": random.randint(2024, 2025),

            # -----------------------------
            # Demographics
            # -----------------------------
            "birth_year": random.randint(1960, 2006),
            "gender_identity": random.choice(genders),
            "country": random.choice(countries),
            "employment_status": random.choice(employment_statuses),
            "work_mode": random.choice(work_modes),

            # -----------------------------
            # Physical attributes (sparse)
            # -----------------------------
            "height_cm": random.choice([round(random.uniform(145, 190), 1), np.nan]),
            "weight_kg": random.choice([round(random.uniform(40, 100), 1), np.nan]),
        }

        # -----------------------------
        # EWA scale (7 items, Likert 0–3)
        # -----------------------------
        for i in range(1, 8):
            row[f"ewa_q{i}"] = (
                random.choice([0, 1, 2, 3])
                if is_completer
                else (np.nan if random.random() < 0.9 else random.choice([0, 1, 2, 3]))
            )

        # -----------------------------
        # UCLA loneliness scale (1–3)
        # -----------------------------
        for i in range(1, 4):
            row[f"ucla_q{i}"] = random.choice([1, 2, 3]) if is_completer else np.nan

        # -----------------------------
        # HPQ productivity scale (1–10)
        # -----------------------------
        for i in range(1, 4):
            row[f"hpq_q{i}"] = random.randint(1, 10) if is_completer else np.nan

        # -----------------------------
        # Derived / summary fields
        # -----------------------------
        row["overall_risk_tier"] = random.choice(risk_tiers) if is_completer else np.nan
        row["work_impact_level"] = random.choice(["Low", "Moderate", "High"]) if is_completer else np.nan
        row["stress_score"] = random.randint(0, 21) if is_completer else np.nan
        row["anxiety_score"] = random.randint(0, 21) if is_completer else np.nan
        row["depression_score"] = random.randint(0, 21) if is_completer else np.nan
        row["loneliness_score"] = random.randint(3, 9) if is_completer else np.nan

        # -----------------------------
        # Text fields (very sparse)
        # -----------------------------
        row["automated_summary_text"] = (
            random.choice([
                "Elevated work-related stress observed.",
                "No significant mental health risks detected.",
                "Moderate emotional strain linked to workload.",
                "Signs of social isolation present."
            ])
            if random.random() < 0.15
            else np.nan
        )

        data.append(row)

    return pd.DataFrame(data)


# -----------------------------
# Generate dataset
# -----------------------------
df_synthetic = generate_public_synthetic_dataset(rows=15000)

# Save for GitHub
df_synthetic.to_csv("synthetic_public_flattened_15k.csv", index=False)
```

### 1. Identify and Map All Raw Items

Based on the column names, here are the items we'll use for scoring.

| Assessment | Hypothesis Variable | Item Columns (String Responses) | Count |
| :--- | :--- | :--- | :--- |
| **Anxiety** (GAD-2) | `Total_Anxiety` | `ewa_q1`, `ewa_q2` | 2 |
| **Depression** (PHQ-2) | `Total_Depression` | `ewa_q3`, `ewa_q4` | 2 |
| **Stress** (DASS-9) | `Total_Stress` | `ewa_q5`, `ewa_q6`, `ewa_q7` | 3 (NOTE: Only 3 DASS-Stress items found. We must assume these 3 represent the DASS-9 Stress items.) |
| **Loneliness** (UCLA-3) | `Total_Loneliness` | `ucla_q1`, `ucla_q2`, `ucla_q3` | 3 |
| **Performance** (HWPQ-3) | `Total_Performance` | `hpq_q1`, `hpq_q2`, `hpq_q3` | 3 (NOTE: These appear to be different types of HWPQ-3 items. We need to check if they require reverse scoring.) |
| **Working Adults** | `working_adults` | `employment_status`- ('Freelance','Full-time', 'Part-time') | 1 |

Since we are missing the full set of 9 DASS stress items, we'll proceed with the three available items and note the limitation. The UCLA-3 items look like they all measure loneliness directly, so we will try direct scoring first, which is simpler and often used in ultra-brief scales. The HWPQ items are on a different numerical scale (6, 4, 5 in the head), which means we cannot use the 0-3 mapping. We'll stick to the raw scores provided.

### 2. Prepare the Code for Scoring and Conversion

```python
# Column groups
gad_cols = ['ewa_q1', 'ewa_q2']
phq_cols = ['ewa_q3', 'ewa_q4']
dass_cols = ['ewa_q5', 'ewa_q6', 'ewa_q7']
ucla_cols = ['ucla_q1', 'ucla_q2', 'ucla_q3']

# Create _score columns
for c in gad_cols + phq_cols + dass_cols:
    df_synthetic[c + "_score"] = pd.to_numeric(df_synthetic[c], errors='coerce')  # already 0-3

for c in ucla_cols:
    df_synthetic[c + "_score"] = pd.to_numeric(df_synthetic[c], errors='coerce') - 1  # 1-3 → 0-2
```

```python
# --- Convert EWA items (GAD, PHQ, DASS) to numeric (already 0–3 in dataset) ---
df_synthetic[gad_cols + phq_cols + dass_cols] = df_synthetic[gad_cols + phq_cols + dass_cols].apply(
    pd.to_numeric, errors='coerce'
)

# --- UCLA items: use your mapped _score columns ---
# Ensure UCLA scores are numeric
for c in ucla_cols:
    df_synthetic[c + "_score"] = pd.to_numeric(df_synthetic[c + "_score"], errors='coerce')

# --- 1. Total Anxiety (GAD-2) ---
df_synthetic['Total_Anxiety'] = df_synthetic['ewa_q1'] + df_synthetic['ewa_q2']

# --- 2. Total Depression (PHQ-2) ---
df_synthetic['Total_Depression'] = df_synthetic['ewa_q3'] + df_synthetic['ewa_q4']

# --- 3. Total Stress (DASS-S) ---
df_synthetic['Total_Stress'] = df_synthetic[['ewa_q5', 'ewa_q6', 'ewa_q7']].sum(axis=1)

# --- 4. Total Loneliness (UCLA-3) ---
df_synthetic['Total_Loneliness'] = df_synthetic[[c + "_score" for c in ucla_cols]].sum(axis=1)

# --- 5. Loneliness Category ---
def loneliness_level(total):
    if total <= 2:
        return 'Low'
    elif total <= 4:
        return 'Moderate'
    else:
        return 'High'

df_synthetic['Loneliness_Level'] = df_synthetic['Total_Loneliness'].apply(loneliness_level)

# --- 6. Total Performance (HWPQ-3) ---
df_synthetic['Total_Performance'] = (
    df_synthetic['hpq_q1'] + df_synthetic['hpq_q2'] - df_synthetic['hpq_q3']
)

# --- 7. Age ---
from datetime import date
current_year = date.today().year
df_synthetic['Age'] = current_year - df_synthetic['birth_year']

# --- 8. Age Group ---
median_age = 35.5
df_synthetic['Age_Group'] = df_synthetic['Age'].apply(
    lambda x: 'Older (> 35.5)' if x > median_age else 'Younger (<= 35.5)'
)

# Fix gender NaN
df_synthetic['gender_identity'] = df_synthetic['gender_identity'].fillna('Not Disclosed')

# --- 9. Clean dataset ---
analysis_vars = [
    'Total_Anxiety',
    'Total_Depression',
    'Total_Stress',
    'Total_Loneliness',
    'Total_Performance',
    'Age'
]

df_clean = df_synthetic.dropna(subset=analysis_vars)

print(f"Original rows: {len(df_synthetic)}. Rows after cleaning: {len(df_clean)}")
print("\nFinal Cleaned Data Sample (Numerical Scores):")
print(df_clean[analysis_vars].head().to_markdown(index=False, numalign='left', stralign='left'))
```
